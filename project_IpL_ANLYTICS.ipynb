3) 10 Best Bowling Figure

### lets extract the best bowling figure on the basis of how much wicket they have taken

```python
# the thing is the wicket is present in column name"best_bowling_figure" befor symbol"--"
df["higher_wickets"]=df['best_bowling_figure'].apply(lambda x:x.split("--")[0])
# convert the string value into "integer"
df["higher_wickets"]=df["higher_wickets"].astype("int")
# let find the bowler name who take more wicket( top 10)
top_bowler = df.groupby("best_bowling")["higher_wickets"].sum().sort_values(ascending = False).head(10)
sns.barplot(x=top_bowler.values,y=top_bowler.index,palette="mako")
```

# **CUSTOM QUESTION AND INSIGHT**

Q1) Who Won The Highest Match By Run

```python
df.head(5)
```

```python
win_by_run = df[df["won_by"] == "Runs"]
biggest_win = win_by_run.loc[win_by_run["margin"].idxmax()]

print(biggest_win["match_winner"], biggest_win["margin"])

```

***Q1.1) Which team has the highest run-margin win***

```python
win_by_run = df[df["won_by"] == "Runs"]

# For each team, take max margin
team_max_margins = win_by_run.groupby("match_winner")["margin"].max()

# Sort descending
team_max_margins = team_max_margins.sort_values(ascending=False)

team_max_margins
```

q2) Wich player has the highest individual score?

```python
# i also use idmax() here too which give me the maximum value row index
high = df.loc[df["highscore"].idxmax()]
print(" THE PLAYER WHO SCORE HIGHEST : ", high["top_scorer"] )
sns.barplot(data=df , y= "top_scorer" , x = "highscore" , palette="rainbow")
```

q3) which bowler has the best bowling figure?

```python
best_bowler_figure = df.loc[df["higher_wickets"].idxmax()]
print(" THE BOWLER WITH THE BEST FIGURE IS : ", best_bowler_figure["best_bowling"] )
plt.figure(figsize=(30,5))
sns.barplot(data=df , y= "best_bowling" , x = "higher_wickets" , palette="rainbow")
```

```python
# finding top 10 best bowler on the basis of high wicket
top10 = df.sort_values(by="higher_wickets", ascending=False).head(10)

print(top10[["best_bowling", "best_bowling_figure", "higher_wickets"]])
```
